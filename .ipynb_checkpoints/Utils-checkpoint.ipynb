```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import write
```

```python
SAMPLE_RATE = 44100 # 1 second is divided into 44100 parts
AMPLITUDE = 4096
```

```python
def get_wave(frequency, duration = 0.5):
    t = np.linspace(0, duration, int(SAMPLE_RATE*duration))
    wave = AMPLITUDE * np.sin(2* np.pi* frequency* t) # A sin(2*pi*f*t)
    return wave

frequency = 440
duration = 1

waveA = get_wave(frequency, duration)
```

```python
plt.plot(waveA[0: int(SAMPLE_RATE/frequency)]) # time for one cycle = SAMPLE_RATE/frequency
plt.xlabel("Time")
plt.ylabel("Amplitude")
plt.show()
```

```python
def get_piano_notes():
    octave = ["C", "c", "D", "d", "E", "F", "f", "G", "g", "A", "a", "B"]
    base_frequency = 261.63 # Middle C
    
    # note_freq = base_frequency * 2^(n/12)
    # For example, if we consider A (n=0) as our base note, then A# will have the frequency = base_freq * 2^(1/12)
    note_frequencies = {octave[i]: base_frequency * pow(2, i/12) for i in range(len(octave))} 
    note_frequencies[''] = 0.0 # silence
    return note_frequencies

get_piano_notes()
```

```python
def get_song_data(song_notes):
    note_frequencies = get_piano_notes()
    song = [get_wave(note_frequencies[note]) for note in song_notes.split("-")]
    song = np.concatenate(song)
    return song
```

```python
song_notes = "C-C-G-G-A-A-G--F-F-E-E-D-D-C--G-G-F-F-E-E-D--G-G-F-F-E-E-D--C-C-G-G-A-A-G--F-F-E-E-D-D-C"
data = get_song_data(song_notes)
data = data * (16300/np.max(data)) # Adjusting the Amplitude (Optional)

plt.plot(data[0:int(SAMPLE_RATE/ 440)])
plt.show()
```

```python
write('twinkle-twinkle.wav', SAMPLE_RATE, data.astype(np.int16))
```

```python
def apply_overtones(frequency, duration, factor, sample_rate=44100, amplitude=4096):
    # factor = amplitudes of the overtones (is a fraction of the original overtone)
    
    #assert abs(1-sum(factor)) < 1e-8
    
    frequencies = np.minimum(np.array([frequency* (x+1) for x in range(len(factor))]), sample_rate//2) # frequencies of overtones
    amplitudes = np.array([amplitude*x for x in factor]) # amplitudes of overtones
    
    fundamental = get_sine_wave(frequencies[0], duration, sample_rate, amplitudes[0]) #sine wave of the fundamental note
    
    for i in range(1, len(factor)):
        overtone = get_sine_wave(frequencies[i], duration, sample_rate, amplitudes[i])
        fundamental += overtone
    return fundamental
    
#apply_overtones(440, 10, [0.5, 0.5, 0.25], 44100, 4096)
```

```python
def get_adsr_weights(frequency, duration, length, decay, sustain_level, sample_rate=44100):
    # ADSR = Attack, Decay, Sustain, and Release
    # When we press an actual piano key, it started out light before quickly getting louder, and the sound 
    # diminishes over time. One model to describe how the sound changes is ADSR (attack, decay, sustain and release).
    # Essentially it describes sound as going through four stages: the initial stage of incline, then descending to
    # a lower level, maintaining there for a little while before diminishing to zero.
    
    # sustain_level = Amplitude of `S` stage as a fraction of max amplitude.
    # length - List of fractions that indicates length of each stage in ADSR.
    # decay - List of fractions that indicates decay factors of each stage
    
    # assert abs(sum(length)-1) < 1e-8
    # assert len(length) ==len(decay) == 4
    
    intervals = int(duration*frequency)
    len_A = np.maximum(int(intervals*length[0]),1)
    len_D = np.maximum(int(intervals*length[1]),1)
    len_S = np.maximum(int(intervals*length[2]),1)
    len_R = np.maximum(int(intervals*length[3]),1)
    
    decay_A = decay[0]
    decay_D = decay[1]
    decay_S = decay[2]
    decay_R = decay[3]
    
    A = 1/np.array([(1-decay_A)**n for n in range(len_A)])
    A = A/np.nanmax(A)
    
    D = np.array([(1-decay_D)**n for n in range(len_D)])
    D = D*(1-sustain_level)+sustain_level
    
    S = np.array([(1-decay_S)**n for n in range(len_S)])
    S = S*sustain_level
    
    R = np.array([(1-decay_R)**n for n in range(len_R)])
    R = R*S[-1]
    
    weights = np.concatenate((A,D,S,R)) # Join a sequence of arrays along an existing axis.
    smoothing = np.array([0.1*(1-0.1)**n for n in range(5)])
    smoothing = smoothing/np.nansum(smoothing) # nansum = convert Nan values to zero by default
    weights = np.convolve(weights, smoothing, mode='same') #  convolve - linear recoil of two 1D sequences.
    
    weights = np.repeat(weights, int(sample_rate*duration/intervals))
    tail = int(sample_rate*duration-weights.shape[0])
    if tail > 0:
        weights = np.concatenate((weights, weights[-1]-weights[-1]/tail*np.arange(tail)))
    return weights
```

```python
def apply_pedal(note_values, bar_value):
    # assert sum(note_values) % bar_value == 0
    new_values = []
    start = 0
    while True:
        cum_value = np.cumsum(np.array(note_values[start:])) # cumsum - cumulative sum of elements
        end = np.where(cum_value == bar_value)[0][0] # where - return element indicating whether true or false
        if end == 0:
            new_values += [note_values[start]]
        else:
            this_bar = np.array(note_values[start:start+end+1])
            new_values += [bar_value-np.sum(this_bar[:i]) for i in range(len(this_bar))]
        start += end+1
        if start == len(note_values):
            break
    return new_values
```
